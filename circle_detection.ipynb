## Circle detection using Opencv's Houghcircles functionality

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
image=cv2.imread('circle.png')
image1= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(image1,cmap='gray')
```

```python
circle = cv2.HoughCircles(image1, cv2.HOUGH_GRADIENT, 1, 260, param1=30, param2=65, minRadius=0, maxRadius=0)
circle=np.round(circle[0, :]).astype("int")
cv2.circle(image,(circle[0][0], circle[0][1]), circle[0][2],(255,0,0), 1) 
plt.imshow(image,cmap='gray')
plt.show()
```

```python
image1=cv2.imread('2.jpg')
gray1= cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)

plt.imshow(gray1,cmap='gray')
```

```python
gray1 = cv2.GaussianBlur(gray1,(5,5),0);
gray1= cv2.medianBlur(gray1,5)
gray1= cv2.adaptiveThreshold(gray1,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,11,2)
kernel = np.ones((5,5), np.uint8)
gray1= cv2.erode(gray1,kernel,iterations=1)
gray1=cv2.dilate(gray1,kernel,iterations=1)
# gray1=cv2.Canny(gray1,100,200)

plt.imshow(gray1,cmap='gray')
```

```python
circles = cv2.HoughCircles(gray1, cv2.HOUGH_GRADIENT, 1, 200, param1=30, param2=45, minRadius=0, maxRadius=0)
print(circles)
if circles is not None:
    circles = np.round(circles[0, :]).astype("int")

    for (x, y, r) in circles:
        cv2.circle(gray1, (x, y), r, (255,255,255), 4)
plt.imshow(gray1,cmap='gray')
plt.show()
```
